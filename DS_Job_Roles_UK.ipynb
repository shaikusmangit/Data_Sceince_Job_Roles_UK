```python
# Replacing pound sign '£' to ' ' And Replacing 'k' to '000'

df['Salary']=df['Salary'].str.replace("£",'').str.replace("K","000")
df
```

```python
df.head(5)
```

```python
# Replacing (Employer est.) to '' AND (Glassdoor est) to ''.

#df['Salary']=df['Salary'].str.replace(r'\(Employer est.\)',"",regex=True)
#df['Salary']=df['Salary'].str.replace(r'\(Glasdoor est.\)',"",regex=True)

# OR..........One line code........
df['Salary']=df['Salary'].str.replace(r'\(Employer est.\)',"",regex=True).str.replace(r'\(Glassdoor est.\)',"",regex=True)
```

```python
df.head(5)
```

```python
# import regex module....
import re
```

```python
# write a function to calculate per hour salary to yearly salary.

def convert_to_yearly(salary):
  if "Per Hour" in str(salary):
    hourly_rate=re.findall(r'\d+\.?\d*',salary)
    if hourly_rate:
      return float(hourly_rate[0])*2080 # converting to yearly salary
  else:
    return salary

# applying the function to dataframe df
df["Salary"]=df["Salary"].apply(convert_to_yearly)
```

```python
df.head()
```

```python
# converting salary column dtype to str

df["Salary"]=df["Salary"].astype(str)
```

```python
#creating 2 new columns as salary_min & salary_max

df[["salary_min","salary_max"]]=df["Salary"].str.split("-",expand=True)
```

```python
df.head()
```

```python
# Removing the white spaces in salary_min & salry _max columns (white spaces before & after -)

df["salary_min"]=df["salary_min"].str.strip()
df["salary_max"]=df["salary_max"].str.strip()
```

```python
# Removing the None values from salary_min & salry_max columns

df["salary_min"]=pd.to_numeric(df["salary_min"],errors="coerce")
df["salary_max"]=pd.to_numeric(df["salary_max"],errors="coerce")
```

```python
# Ensuring location column dtype is str type

df["Location"]=df["Location"].astype(str)
```

```python
# Split location column into 'City', 'Country' based on , comma delimiter

df[["City","Country"]]=df["Location"].str.split(",",n=1,expand=True)
```

```python
df.head(5)
```

```python
# Using Onehot encoding (i.e 0 and 1) if skill is Yes then 1, No 0 will be encoded to all skills......AND split column skills by , comma
skills_split=df["Skills"].str.get_dummies(sep=",")
```

```python
skills_split.head()
```

```python
# Combining the Two data Frames i.e, new data frame name df_with_skills ==> old dataframes=(df+skills_split)

df_with_skills=pd.concat([df,skills_split],axis=1)
```

```python
# new dataframe df_with_skills

df_with_skills.head()
```

```python
# Creating new variable skill_frequencies to sum the data of all skills (i.e 1 and 0)
# iloc helps us to locate the column
# : represetns find the all the rows
# I want all the columns starting from column 'AI' upto the End so using : in end
# using sum() to sum the all values

skill_frequencies=df_with_skills.iloc[:,df_with_skills.columns.get_loc('AI'):].sum()
```

```python
skill_frequencies
```

```python
# Import world cloud module

from wordcloud import WordCloud
```

```python
#create wordcloud

# with width,height & bacground white.

#.generate_from_frequencies function here pass our variable skill_frequencies.

wordcloud=WordCloud(width=800,height=400,background_color='white').generate_from_frequencies(skill_frequencies)

plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.title('Most popular skills word cloud',fontsize=16)
plt.show()
```

```python
# Creating a new variable skills_column, then using ilco : to select all skills till the end,
# and sorting the data in decending order i.e Top skills will come first.

skills_column=df_with_skills.iloc[:,df_with_skills.columns.get_loc('AI'):].sum().sort_values(ascending=False)
```

```python
# creating new variable and Selecting Top 10 skills

top_10_skills=skills_column.head(10)
```

```python
# Plotting the data in Vertical bar chart

plt.figure(figsize=(10,6))
sns.barplot(x=top_10_skills.index,y=top_10_skills.values)
plt.xlabel('Skills')
plt.ylabel('Frequency')
plt.title('Top 10 Skills')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting the data in bar chart Horizontally using (barh)

plt.figure(figsize=(10,6))
top_10_skills.plot(kind='barh',color='green') #kind=barh means bar chart h= horizontally.
plt.xlabel('Number of job Listings',fontsize=12)
plt.ylabel('Skills',fontsize=12)
plt.title('Top 10 Skills',fontsize=16)
plt.gca().invert_yaxis()
plt.show()
```
